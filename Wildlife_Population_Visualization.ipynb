```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

filename = '/content/drive/MyDrive/Coding Dojo/Week 4/Wildlife Population Change.xlsx'
df = pd.read_excel(filename)

df
```

```python
animal_groups = df['Taxonomic Group'].values
```

```python
population_decrease = df['Decreasing Populations'].values
```

```python
population_increase = df['Increasing Populations'].values
```

```python
# The plot below shows animal population increase. 

plt.style.use('seaborn')
plt.plot(animal_groups, population_increase,c='r',marker = '.', markersize = 10)
plt.xlabel('Taxonomic Group',fontsize =16)
plt.ylabel('Increasing Populations',fontsize =16)
plt.title('Number of Animal Increase', fontsize =16);
```

```python
# The visualization below shows both the increasing and decreasing populations of the Taxonomic Group

plt.plot(animal_groups, population_increase, c = 'r', label = 'Increasing Populations')
plt.scatter(animal_groups, population_decrease, c= 'k', label = 'Decreasing Populations')
plt.legend(loc=(1,1))
plt.xlabel('Taxonomic Group', fontsize = 15)
plt.ylabel('Population Change in %', fontsize = 15)
plt.title('Animal Population Change', fontsize = 25)
plt.ylim(bottom=0.35,top=0.60)
plt.xticks(rotation = 90);
plt.grid(c = 'b', 
         alpha = .9,
         linestyle = '-')
```
